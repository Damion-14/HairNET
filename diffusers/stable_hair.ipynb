```python
# Ensure you are in the right directory
%cd /content/Stable-Hair

# Install necessary libraries (add others if needed)
!pip install -q omegaconf gradio torch diffusers accelerate bitsandbytes transformers einops opencv-python Pillow

# Imports needed for the setup
import torch
import gc
from omegaconf import OmegaConf
import os
from PIL import Image
import numpy as np
import cv2 # Ensure opencv is installed

# Import classes from the Stable-Hair repository files
from diffusers import DDIMScheduler, UniPCMultistepScheduler
from diffusers.models import UNet2DConditionModel
from ref_encoder.latent_controlnet import ControlNetModel # Assuming these paths are correct relative to Stable-Hair
from ref_encoder.adapter import *
from ref_encoder.reference_unet import ref_unet
from utils.pipeline import StableHairPipeline
from utils.pipeline_cn import StableDiffusionControlNetPipeline

import gradio as gr

# Define essential variables
device = "cuda" if torch.cuda.is_available() else "cpu"
# Use float16 as the default for non-quantized parts or fallbacks
# We'll try to load base models in 8-bit where possible
dtype_fallback = torch.float16
config_path = "./configs/hair_transfer.yaml"
config = OmegaConf.load(config_path)

print(f"Using device: {device}")
print("Libraries and config loaded.")
```

```python
import torch
import numpy as np
from PIL import Image

def perform_hair_transfer(source_image_np, reference_image_np, random_seed=-1, step=30, guidance_scale=1.5, scale=1.0, controlnet_conditioning_scale=1.0):
    """Replicates Hair_Transfer logic using loaded components."""
    prompt = ""
    n_prompt = ""
    random_seed = int(random_seed)
    step = int(step)
    guidance_scale = float(guidance_scale)
    scale = float(scale) # Hair encoder scale
    controlnet_conditioning_scale = float(controlnet_conditioning_scale) # Latent identity net scale

    # Assuming source_image_np and reference_image_np are numpy arrays HxWxC (0-255)
    H, W, C = source_image_np.shape

    print("Performing hair transfer...")
    # Set the scale for the hair adapter
    set_scale(main_pipeline.unet, scale) # Requires set_scale func from adapter.py

    generator = torch.Generator(device=main_pipeline.device)
    if random_seed != -1:
        generator.manual_seed(random_seed)
    else:
        generator.seed() # Use a truly random seed if -1

    with torch.no_grad(): # Important for inference
         # Requesting "pt" might be ignored by custom pipeline, but doesn't hurt to leave it
         output = main_pipeline(
             prompt,
             negative_prompt=n_prompt,
             num_inference_steps=step,
             guidance_scale=guidance_scale,
             width=W,
             height=H,
             controlnet_condition=source_image_np,
             controlnet_conditioning_scale=controlnet_conditioning_scale,
             generator=generator,
             reference_encoder=hair_encoder,
             ref_image=reference_image_np,
             output_type = "pt"
         )

    print("Hair transfer complete.")

    if hasattr(output, 'samples'):
        print("DEBUG: Found 'samples' attribute.")
        if isinstance(output.samples, np.ndarray):
            print("DEBUG: output.samples is a numpy.ndarray.")
            sample_np = output.samples # This is the numpy array.
            # Convert the numpy array into a PIL image
            sample_np = (sample_np.clip(0, 1) * 255).astype(np.uint8)
            sample_pil = Image.fromarray(sample_np)
            return sample_pil # Return the PIL image.
        else:
            print(f"DEBUG: output.samples is not a numpy.ndarray, but a {type(output.samples)}")
            return np.zeros_like(source_image_np) # Return a black image.
    else:
        print("DEBUG: Did not find 'samples' attribute in output.")
        return np.zeros_like(source_image_np) # Return a black image.

def perform_get_bald(id_image_pil, scale=1.0):
    """Replicates get_bald logic using loaded components."""
    print("Performing hair removal...")
    H, W = id_image_pil.size
    scale = float(scale)

    with torch.no_grad(): # Important for inference
        image = remove_hair_pipeline(
            prompt="", # Empty prompts as in original code
            negative_prompt="",
            num_inference_steps=30,
            guidance_scale=1.5,
            width=W,
            height=H,
            image=id_image_pil, # Pipeline likely expects PIL image
            controlnet_conditioning_scale=scale,
            generator=None, # Or create a generator if needed
        ).images[0] # Get the first PIL image

    print("Hair removal complete.")
    return image # Returns PIL image

def model_call_notebook(id_image_np, ref_hair_np, converter_scale, scale, guidance_scale, controlnet_conditioning_scale):
    """Gradio callback using notebook functions and loaded components."""
    print("Received data in Gradio callback.")
    # Input images are numpy arrays from Gradio Image input
    id_image = Image.fromarray(id_image_np.astype('uint8'), 'RGB')
    ref_hair = Image.fromarray(ref_hair_np.astype('uint8'), 'RGB')

    # Resize images
    id_image = id_image.resize((512, 512))
    ref_hair = ref_hair.resize((512, 512))
    print("Images resized to 512x512.")

    # Step 1: Get bald image
    id_image_bald_pil = perform_get_bald(id_image, converter_scale)
    id_image_bald_np = np.array(id_image_bald_pil)

    # Step 2: Hair Transfer
    transfer_result = perform_hair_transfer(
        source_image_np=id_image_bald_np,
        reference_image_np=ref_hair_np,
        random_seed=-1,
        step=30,
        guidance_scale=guidance_scale,
        scale=scale,
        controlnet_conditioning_scale=controlnet_conditioning_scale
    )

    # Check if transfer_result is a PIL Image
    if isinstance(transfer_result, Image.Image):
        # Convert PIL Image back to NumPy array
        transfer_result_np = np.array(transfer_result)
    else:
        # If it's not a PIL Image (e.g., an error image), use it directly
        transfer_result_np = transfer_result

    # Gradio output expects PIL images
    transfer_result_pil = Image.fromarray(transfer_result_np)
    print("Processing complete. Returning images to Gradio.")

    return id_image_bald_pil, transfer_result_pil
```

```python
# --- Component 1: Base UNet (Attempt 8-bit) ---
print("Loading Base UNet...")
try:
    unet_base = UNet2DConditionModel.from_pretrained(
        config.pretrained_model_path,
        subfolder="unet",
        load_in_8bit=True,  # <<< Apply bitsandbytes here
        device_map="auto"   # <<< Use Accelerate for device mapping
    )
    print("Base UNet loaded in 8-bit.")
    # Note: device_map might place parts on CPU. Careful casting needed later.
except Exception as e:
    print(f"WARNING: Failed to load base UNet in 8-bit: {e}. Falling back to {dtype_fallback}.")
    unet_base = UNet2DConditionModel.from_pretrained(
        config.pretrained_model_path, subfolder="unet"
    ).to(device).to(dtype_fallback)
gc.collect()
torch.cuda.empty_cache()

# --- Component 2: ControlNet for Main Pipeline ---
print("Loading Main ControlNet...")
# Needs the unet on CPU for init if device_map was used, or ensure it's on the target device
unet_for_cn_init = unet_base.to("cpu") if hasattr(unet_base, 'hf_device_map') else unet_base.to(device)
controlnet_main = ControlNetModel.from_unet(unet_for_cn_init).to(device)
print("  Loading ControlNet state dict...")
_state_dict = torch.load(os.path.join(config.pretrained_folder, config.controlnet_path), weights_only=True, map_location="cpu")
controlnet_main.load_state_dict(_state_dict, strict=False)
controlnet_main = controlnet_main.to(dtype_fallback) # Cast weights loaded from .bin to fp16
del _state_dict
print("Main ControlNet loaded and cast to", dtype_fallback)
gc.collect()
torch.cuda.empty_cache()

# --- Component 3: Main StableHair Pipeline (Attempt 8-bit base) ---
print("Loading Main StableHair Pipeline...")
try:
    # Don't reload unet/controlnet if already loaded. Pass controlnet.
    # Pipeline will load VAE, text_encoder etc. Try 8-bit for those.
    main_pipeline = StableHairPipeline.from_pretrained(
        config.pretrained_model_path,
        controlnet=controlnet_main, # Pass pre-loaded ControlNet
        safety_checker=None,
        # torch_dtype=dtype_fallback, # Omit when using load_in_8bit
        load_in_8bit=True,       # <<< Apply bitsandbytes to other components (VAE, Text Enc)
        device_map="auto"
    )
    # Ensure controlnet is on the same device as pipeline components after device_map
    main_pipeline.controlnet = controlnet_main.to(main_pipeline.device)
    print("Main Pipeline base components loaded in 8-bit.")
except Exception as e:
    print(f"WARNING: Failed to load main pipeline in 8-bit: {e}. Falling back to {dtype_fallback}.")
    main_pipeline = StableHairPipeline.from_pretrained(
        config.pretrained_model_path,
        controlnet=controlnet_main,
        safety_checker=None,
        torch_dtype=dtype_fallback,
    ).to(device)
main_pipeline.scheduler = DDIMScheduler.from_config(main_pipeline.scheduler.config)
print("Main StableHair Pipeline loaded.")
gc.collect()
torch.cuda.empty_cache()

# --- Component 4: Hair Encoder (ref_unet) ---
print("Loading Hair Encoder (ref_unet)...")
# Initialize structure (could try 8-bit base here too, but loading float weights after is tricky)
# Sticking to float16 base for simplicity when loading float weights after
hair_encoder = ref_unet.from_pretrained(config.pretrained_model_path, subfolder="unet").to(device)
print("  Loading Hair Encoder state dict...")
_state_dict = torch.load(os.path.join(config.pretrained_folder, config.encoder_path), weights_only=True, map_location="cpu")
hair_encoder.load_state_dict(_state_dict, strict=False)
hair_encoder = hair_encoder.to(dtype_fallback) # Cast weights loaded from .bin to fp16
del _state_dict
print("Hair Encoder loaded and cast to", dtype_fallback)
gc.collect()
torch.cuda.empty_cache()

# --- Component 5: Hair Adapter ---
print("Loading Hair Adapter...")
# Adapter injection seems tied to the main pipeline's UNet
# Use the UNet instance from the loaded main_pipeline
hair_adapter = adapter_injection(main_pipeline.unet, device=main_pipeline.device, dtype=dtype_fallback, use_resampler=False)
print("  Loading Hair Adapter state dict...")
_state_dict = torch.load(os.path.join(config.pretrained_folder, config.adapter_path), weights_only=True, map_location="cpu")
hair_adapter.load_state_dict(_state_dict, strict=False)
# Adapter might already be float16 internally, but ensure consistency
hair_adapter = hair_adapter.to(dtype_fallback)
del _state_dict
print("Hair Adapter loaded and cast to", dtype_fallback)
gc.collect()
torch.cuda.empty_cache()

# --- Component 6: Bald Converter ControlNet ---
print("Loading Bald Converter ControlNet...")
# Re-use the base unet structure if possible, ensures consistency
unet_for_bald_init = unet_base.to("cpu") if hasattr(unet_base, 'hf_device_map') else unet_base.to(device)
bald_converter = ControlNetModel.from_unet(unet_for_bald_init).to(device)
unet_for_bald_init.to('meta') # Free memory
print("  Loading Bald Converter state dict...")
_state_dict = torch.load(config.bald_converter_path, weights_only=True, map_location="cpu")
bald_converter.load_state_dict(_state_dict, strict=False)
bald_converter = bald_converter.to(dtype_fallback) # Cast weights loaded from .bin to fp16
del _state_dict
# We can now likely delete the base UNet if it's not needed elsewhere
del unet_base
print("Bald Converter loaded and cast to", dtype_fallback)
gc.collect()
torch.cuda.empty_cache()

# --- Component 7: Hair Removal Pipeline (Attempt 8-bit base) ---
print("Loading Hair Removal Pipeline...")
try:
    remove_hair_pipeline = StableDiffusionControlNetPipeline.from_pretrained(
        config.pretrained_model_path,
        controlnet=bald_converter, # Pass pre-loaded ControlNet
        safety_checker=None,
        # torch_dtype=dtype_fallback, # Omit when using load_in_8bit
        load_in_8bit=True, # << using 8bit from Bits and Bytes
        device_map="auto"
    )
    # Ensure controlnet is on the right device
    remove_hair_pipeline.controlnet = bald_converter.to(remove_hair_pipeline.device)
    print("Hair Removal Pipeline base components loaded in 8-bit.")
except Exception as e:
    print(f"WARNING: Failed to load Hair Removal pipeline in 8-bit: {e}. Falling back to {dtype_fallback}.")
    remove_hair_pipeline = StableDiffusionControlNetPipeline.from_pretrained(
        config.pretrained_model_path,
        controlnet=bald_converter,
        safety_checker=None,
        torch_dtype=dtype_fallback,
    ).to(device)
remove_hair_pipeline.scheduler = UniPCMultistepScheduler.from_config(remove_hair_pipeline.scheduler.config)
print("Hair Removal Pipeline loaded.")
gc.collect()
torch.cuda.empty_cache()

print("\n--- All components loaded ---")
```

```python
print("Setting up Gradio interface...")

# Use gr.Image which provides numpy arrays by default
image1 = gr.Image(label="id_image", type="numpy")
image2 = gr.Image(label="ref_hair", type="numpy")
number0 = gr.Slider(minimum=0.5, maximum=1.5, value=1, label="Converter Scale")
number1 = gr.Slider(minimum=0.0, maximum=3, value=1.0, label="Hair Encoder Scale")
number2 = gr.Slider(minimum=1.1, maximum=3.0, value=1.5, label="CFG (Guidance Scale)")
number3 = gr.Slider(minimum=0.1, maximum=2.0, value=1, label="Latent IdentityNet Scale")
# Output components expect PIL images if our callback returns them
output1 = gr.Image(type="pil", label="Bald_Result")
output2 = gr.Image(type="pil", label="Transfer Result")

iface = gr.Interface(
    fn=model_call_notebook, # Use the notebook function
    inputs=[image1, image2, number0, number1, number2, number3],
    outputs=[output1, output2],
    title="Hair Transfer Demo (Notebook Quantized)",
    description="Running Stable-Hair with attempted 8-bit quantization for base models directly in notebook. Input images will be resized to 512x512."
)

print("Launching Gradio...")
# Launch on a shareable link if on Colab
iface.queue().launch(share=True, debug=True) # Add debug=True for more logs
```
